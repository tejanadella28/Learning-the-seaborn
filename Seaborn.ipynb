```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
sns.barplot (x='who', y = 'age', data = df)
```

```python
sns.barplot (x='sex', y = 'age', hue = 'class', data = df)
```

```python
sns.barplot (x='class', y = 'survived', hue = 'sex', data = df)
```

```python
df['embark_town'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
sns.distplot(df[df["survived"]==0]["age"], hist = False)
sns.distplot(df[df["survived"]==1]["age"], hist = False)
```

```python
sns.countplot(y='embark_town',data=df)
```

```python
sns.countplot(x='class', hue ='embark_town', data=df)
```
